```python
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
% matplotlib inline
```

```python
viols_file = "SF_Viols.csv"

df_viols = pd.read_csv(viols_file)

df_viols.head()
```

```python
print(f"There were {len(df_viols)} restaurant inspection code violations observed in LA in 2017.")
#len(df_viols)
```

```python
# Start with one review:


stopwords = set(STOPWORDS)
stopwords.update(["in", "for", "if","or", "and", ":"])

# Generate a word cloud image
wordcloud = WordCloud(stopwords=stopwords, width=800, height=500, max_font_size=70, max_words=300,scale=3, background_color="black").generate(' '.join(df_viols['Violations']))
# Create and generate a word cloud image:

```

```python
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wordcloud.to_file("SF_review.png")
```
