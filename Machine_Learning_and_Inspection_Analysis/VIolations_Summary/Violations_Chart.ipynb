```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
df = pd.read_csv("Violations_Summary.csv")
df.head()
df.set_index('Violation')
```

```python
#>>> table = pivot_table(df, values='D', index=['A', 'B'],
#...                     columns=['C'], aggfunc=np.sum)
#

cities = pd.pivot_table(df,values = 'Count',index = 'Violation', columns = 'City', aggfunc=np.sum)
cities = cities.sort_values(by=['LA'], ascending = False)
cities.index.values
cities.head()
```

```python
violations = cities.index.values
y_pos = np.arange(len(violations))
sf_counts = cities['SF']
la_counts = cities['LA']
 
plt.bar(y_pos, sf_counts, align='center', alpha=0.5)
plt.bar(y_pos, la_counts, align='center', alpha=0.5)

plt.xticks(y_pos, violations)
plt.xticks(rotation=90)
plt.ylabel('Usage')
plt.title('Violations by City')
 
plt.show()
```

```python
plt.figure(figsize=(8,10))

violations = cities.index.values
y_pos = np.arange(len(violations))
sf_counts = cities['SF']
la_counts = cities['LA']
 
plt.barh(y_pos, sf_counts, align='center', alpha=0.5, label = 'SF')
plt.barh(y_pos, la_counts, align='center', alpha=0.5, label = 'LA')

plt.yticks(y_pos, violations)
#plt.xticks(y_pos, range(1,4200))
plt.xlabel('Count')
plt.title('Violations by City')
#plt.legend()
plt.legend(prop={'size': 15})

plt.show()
```
